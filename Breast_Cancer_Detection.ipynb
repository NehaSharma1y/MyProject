```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn .datasets
from sklearn.model_selection import train_test_split
```

```python
x=dataframe.drop(columns='label',axis=1)
y=dataframe['label']
```

Splitting the data into training and testing data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(y.shape,y_train.shape,y_test.shape)
```

Standardize the data


```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_train_standard=scaler.fit_transform(x_train)
x_test_standard=scaler.transform(x_test)
```

creating neural network

```python
#tensorflow is used to create neural network
# keras is just like the wrapper of tenserflow here...
#keras use tenserflow as background to create neural network and create some APIs and functions to improve the functionality of neural network
#anytime when we give the data to modekl it does not give accurate or same output...it always give diffrent output....to use random seed it will give same output
#set random seed juct after importing the tensorflow library
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

Setting up the layers of neural network

```python
#Flatten is used for input layer it will conver the data into single dimentional array
#Dense means the beuron of this particular laye is connected with every neuron of previous layer. need to give the number of neurons in the layer and activation function
model= keras.Sequential([keras.layers.Flatten(input_shape=(30,)),
                        keras.layers.Dense(20,activation='relu'),
                        keras.layers.Dense(2,activation='sigmoid')   ])

```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history=model.fit(x_train_standard,y_train,validation_split=0.1,epochs=10)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data','validation data'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data','validation data'],loc='upper right')
```

Accuracy of the model on test data

```python
loss,accuracy=model.evaluate(x_test_standard,y_test)
print(accuracy)
```
